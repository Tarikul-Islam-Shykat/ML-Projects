```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as p
```

Just a line

```python
x  = [1,2,3]
y = [2,4,6]

plt.title('our first graph', fontdict= {'fontname': 'Comic Sans MS', 'fontsize': 20}) # adds titles to our graph.

# label your graph
plt.xlabel("X axis")
plt.ylabel("Y axis") 

plt.plot(x,y)
plt.show()
```

For manullay showing tick marks 

```python
plt.xticks([0,1,2,3,4,5,6])
plt.yticks([0,3,6,8,9])
plt.plot(x,y, label = '2x', color = 'red', linewidth=5, marker = '.', linestyle = '--', markersize = 20 , markeredgecolor = 'blue') #change the color. line width change the width of your line. marker size do the same. 
plt.legend() # this will label you graph 
plt.show()
```

Another way


```python
plt.plot(x,y, 'r^--',label = '2x') #change the color. line width change the width of your line. marker size do the same. 
plt.legend() # this will label you graph 
plt.show()
```

Line number 2

```python
x2 = np.arange(0,3.5,0.5) #the arange() function is used to get evenly spaced values within a given interval.
print(x2, x2**2)
plt.plot(x2, 'r', x2**2, label ='x2**2')

plt.legend()# to show the label
plt.show()
```

resize the graph

```python
plt.figure(figsize= (1,2),  dpi = 200 ) # dpi increase the size of graph
plt.plot(x2, 'r', x2**2, label ='x2**2')

plt.show()
```

Bar chart

```python
labels = ['A','B','C']
values = [1,3,4]
bars = plt.bar(labels, values)

# for individual design of the graph
bars[0].set_hatch('/')
bars[1].set_hatch('o')

plt.show() # you can add the title like the last graph.
```
