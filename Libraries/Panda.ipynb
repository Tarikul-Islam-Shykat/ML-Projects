**Filtering data**

```python
df.loc[df['Type 1']== 'Grass'] # one condition
```

```python
df.loc[(df['Type 1']== 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)] # multiple condition data 
```

# get it as a new data

```python
new_df = df.loc[(df['Type 1']== 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)] 
print(new_df)
```

```python
# you can see from above the index is still the same, if you want change it, then you can easily reset it
new_df = new_df.reset_index(drop = True) # drop usually gets you rid of the old indexes 
print(new_df)
```

# **Regex Filtering **

```python
import re
df.loc[df['Name'].str.contains('Mega')] # gives you the rows which contain Mega in their rows. you can also give regex 
```

```python
df.loc[df['Type 1'].str.contains('fire|Grass', flags = re.I ,regex = True)] # to ignore case use flags = re.I
```

Working With Large amounts of data

```python
for df in pd.read_csv('modified.csv', chunksize = 5): # 5 means five rows are given to this at this time 
  print(df)
```

Concat with large dataset and create a new dataset

```python
new_df = pd.DataFrame(columns = df.columns)
for df in pd.read_csv('modified.csv', chunksize = 5): # 5 means five rows are given to this at this time 
  result = df.groupby(['Type 1']).count()
  new_df = pd.concat([new_df, result])

print(new_df)
```
